## Simulate the a RSF study

Going to simulate an RSF study so I can get a feel for how this is actually working.  Basically, I am going to try to replicate the simulationg results of Aart et al. 2012 MEE.

Steps
-----
1. Simulate a landscape with one environmental covariate
2. Place some number of individual points on this landscape according to a resource selection function
3. Use different RSF strategies to recover the the resource selection function

### Step 1: Simulate the landscape

```python
## Simulate a landscape using a Multivariate
import scipy.stats as stats
from sklearn.metrics.pairwise import pairwise_distances
points = np.arange(0, 60)
X, Y = np.meshgrid(points, points)

# Calculate pairwise distances
pairs = []
for y in points:
    for x in points:
        pairs.append((x, y))
        
dist_mat = pairwise_distances(np.array(pairs))
```

```python
# Compute correlation matrix: K_ij = eta^2 exp(-rho^2 D_ij^2) + delta*sigma
eta = 0.2
rho = 0.6
sigma = 2
K_mat = eta**2 * np.exp(dist_mat**2 * -rho**2)
mu = np.zeros(len(pairs)) # Mean of zero
```

```python
environment = np.random.multivariate_normal(mu, K_mat).reshape(len(points), len(points))
```

```python
plt.pcolormesh(X, Y, environment)
plt.colorbar()
```

### Step 2: Allocated individuals to the landscape

```python
# Allocate individuals to the landscape 
N = 3000
beta1 = 0.5
rsf = lambda x, beta: np.exp(beta*x)
rsf_vals = rsf(environment.flatten(), beta1)
rsf_probs = rsf_vals / np.sum(rsf_vals)
counts = np.random.multinomial(N, rsf_probs)
```

```python
print(np.min(rsf_probs))
print(np.max(rsf_probs))
```

```python
# Plot the occurrence on the landscape
plt.pcolormesh(X, Y, environment)
plt.colorbar()
ms = 2
for i, pair in enumerate(pairs):
    if counts[i] == 1:
        plt.plot(pair[0], pair[1], 'o', ms=ms, color="black")
    if counts[i] == 2:
        plt.plot(pair[0], pair[1], 'o', ms=ms, color="red")
    if counts[i] > 2:
        plt.plot(pair[0], pair[1], 'o', ms=ms, color="grey")
```

```python
# Increase grid size and explore the effects on Poisson GLM and

def blockshaped(arr, nrows, ncols):
    """
    Return an array of shape (n, nrows, ncols) where
    n * nrows * ncols = arr.size

    If arr is a 2D array, the returned array should look like n subblocks with
    each subblock preserving the "physical" layout of arr.
    """
    h, w = arr.shape
    return (arr.reshape(h//nrows, nrows, -1, ncols)
               .swapaxes(1,2)
               .reshape(-1, nrows, ncols))
```
